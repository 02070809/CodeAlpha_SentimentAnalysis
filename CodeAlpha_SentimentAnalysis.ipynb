```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
reviews = [
    "This product is absolutely amazing and works perfectly",
    "Worst purchase I ever made, totally disappointed",
    "Product is okay, nothing special",
    "Excellent quality, very happy with this purchase",
    "Terrible product, broke after one day",
    "It is fine, does the job",
    "Highly recommend this to everyone, fantastic!",
    "Not worth the money at all",
    "Average product, expected better quality",
    "Best product I have ever bought, love it!",
    "Very bad experience, will not buy again",
    "Decent product for the price",
    "Outstanding quality and fast delivery",
    "Completely useless product, waste of money",
    "Good product overall, happy with purchase"
]

df = pd.DataFrame(reviews, columns=['Review'])
df.head()
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['Review'].apply(get_sentiment)
print(df)
```

```python
print("=== SENTIMENT ANALYSIS RESULTS ===")
print(df['Sentiment'].value_counts())
```

```python
plt.figure(figsize=(6,6))
df['Sentiment'].value_counts().plot(kind='pie',
                                     autopct='%1.1f%%',
                                     colors=['green','red','yellow'])
plt.title('Customer Review Sentiment Analysis')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Sentiment'].value_counts().plot(kind='bar',
                                     color=['green','red','yellow'])
plt.title('Sentiment Distribution of Customer Reviews')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()
```
